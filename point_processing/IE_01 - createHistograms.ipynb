## Create Histograms

Histograms basically shows the intensity distribution in an image. It is a graph between the pixel intensities vs the number of pixels. Below is a python program which can take an image and can produce an intensity distribution graph or histogram.

```python
from scipy import misc
import numpy as np
from matplotlib import pyplot as plt
```

Create a 100x100 gray scale image by randomly creating an array with random pixel values ranging from 0 to 255. Then plot the image and its histogram as shown below. To plot the histogram, call the hist method in matplotlib library in python by passing the flattened image array and set the bins to auto.

```python
hist1 = np.zeros([100,100],int)
for i in range(0,100):
    hist1[i,:] = np.random.randint(0,255,100)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(hist1,cmap='gray')

plt.subplot(1,2,2)
plt.hist(hist1.ravel(),bins='auto')
```

   Import two more images present in the skimage library in the data folder (moon, chelsea). Moon image is a gray scale image, a very low contrast and poor quality image. So when histogram is plotted for the moon image, it shows that some particular intensity levels have very high distribution in comparison to others. In chelsea image, it is a RGB image of a cat which is a good quality image. It is converted to a gray scale image first with the formula written below and then histogram is plotted. 

```python
from skimage import data

from matplotlib import pyplot as plt
import numpy as np

moon = data.moon()
chelsea = data.chelsea()

chelsea_gray = np.dot(chelsea[...],[0.299,0.587,0.114])

plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
plt.imshow(moon,cmap='gray')

plt.subplot(2,2,2)
plt.imshow(chelsea_gray,cmap='gray')

plt.subplot(2,2,3)
plt.hist(moon.ravel(),bins='auto')

plt.subplot(2,2,4)
plt.hist(chelsea_gray.ravel(),bins='auto')

plt.show()
```
